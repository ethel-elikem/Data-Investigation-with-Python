### Research Question 3 (What has been the general popularity of movies from year to year?)

```python
poplarity_pattern = nmovie_data.groupby('release_year').sum().reset_index().tail(25)
poplarity_pattern.plot.bar(x='release_year',y='popularity', figsize = (12,8), color='r', xlabel= 'years', ylabel = 'sum popularity', 
                           title ="General popularity of movies over the years")
plt.show()
```

**We would also like to investigate the general popularity of movies over the years, irrespective of genres. We sum the popularity of movies by release year. For visibility, the plot starts from the year $1991$ and we see that the general trend in the popularity of movies is an upward trend. We however see that in some years such as, $2000$, $2010$ and $2012$ there is a dip in the popularity of movies which could be caused by external factors outside the movie industry or space.**

### Research Question 3 (Which genre has the highest popularity?)

```python
temp2 = nmovie_data.groupby('genres').sum().reset_index()
category = {}
for (elm1,elm2) in temp2[['genres','popularity']].values:
    splt = elm1.split('|')
    for elm3 in splt:
        if (elm3 not in category ):
            category[elm3] = elm2
        else:
            category[elm3] += elm2
```

```python
temp3 = nmovie_data.groupby(['genres','release_year']).sum().reset_index()
category= {}
for (elm1,elm2) in temp3[['genres','popularity']].values:
    splt = elm1.split('|')
    for elm3 in splt:
        if (elm3 not in category):
            category[elm3] = elm2
        else:
            category[elm3] += elm2
genpop =pd.DataFrame.from_dict(category, orient='index').reset_index()
genpop.columns=['genres', 'sum_popularity']       
genpop
```

```python
genpop.plot.bar(x='genres',y='sum_popularity', color = 'r', ylabel='sum popularity',figsize=(12,8))
plt.title("sum popularity of genres over the years")
plt.show()
```

**With the third research question, we want to know which movie genres are most popular throughout the years. To do this we extract the individual genres, $20$ of them, and then we sum the popularity of the movies which fall under the these genres. From the bar plot, we see that $drama$ is the most popular genre, followed by $action$, $comedy$ and $thriller$. This would also mean that most of the popular movies fall under at least one of these individual genres. the genre $foreign$ is the least popular followed closely by $tv movie$.**

### Research Question 4 (Which genres are most popular from year to year?)

```python
temp3 = nmovie_data.groupby(['genres','release_year']).sum().reset_index()

category2 = {}
l = {}
for (elm1,elm2,year) in temp3[['genres','popularity','release_year']].values:
    splt = elm1.split('|')
    if year in category2:
        l = category2[year]
    else:
        l={}
    for elm3 in splt:
        if (elm3 not in l):
            l[elm3] = elm2
        else:
            l[elm3] += elm2
    category2[year] = l
```

```python
category2 = dict(sorted(category2.items()))
for year in category2:
    dat = category2[year]
    dat =pd.DataFrame.from_dict(dat, orient='index').reset_index()
    dat.columns=['genres', 'sum_popularity']   
    dat.plot.bar(x='genres',y='sum_popularity', color = 'r', ylabel='sum popularity',title=str(year))
   
```

**Finally we investigate the popularity of genres from year to year. There are a total of $45$ plots one for each year from $1960$ to $2015$.  It can be observed that whilethe popularity of a genre varies from year to year, the popular genres have remained fairly stable over the years. The factors that influenced these varitations in the popularity of theses genres maybe external or within the movie industry and are beyond the scope of this study.**
